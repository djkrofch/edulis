```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy, os, datetime

gcolour=[0.9,0.5, 0.0];
ngcolour=[0.9, 0.6, 0.0];
scolour=[0.6, 0.2, 0];
jcolour=[0.25, 1.0, 0.0];
pjcolour=[0.0, 0.5, 0.0];
pjgcolour=[0.2, 0.5, 0.0];
pcolour=[0.5, 0.5, 1.0];
mcolour=[0.0, 0.0, 0.6];
siteColors = [gcolour, scolour, jcolour, pjcolour, pcolour, mcolour]

dataDir = '/home/nikko/Research/Data/TowerData/AmerifluxFiles/Products/Daily/'
```

```python
# Read in daily files, and make one large dataframe
# Generator function to produce file names
def files(path):
    for file in os.listdir(path):
        if os.path.isfile(os.path.join(path, file)):
            yield file
            
def getSiteName(filename):
    sitename = filename.split('-')[1].split('_')[0]
    return sitename
            
filelist = []
check = files(dataDir)
for names in check:
    filelist.append(names)

dailyDict = []
nextDict = []

count = 0
for idx in arange(0,len(filelist)):
    if count == 0:
        dataDict = pd.read_csv(dataDir + filelist[idx], sep = ',')
        dataDict['SITE'] = getSiteName(filelist[idx])
        count = count + 1
    else:
        nextDict = pd.read_csv(dataDir + filelist[idx], sep = ',')
        nextDict['SITE'] = getSiteName(filelist[idx])
    
        dataDict = dataDict.append(nextDict, ignore_index=True)

     
timestamp = []
idx = 0
for time in arange(0,dataDict.year.size):
    #print time
    idx = idx + 1
    timestamp.append(datetime.datetime(dataDict.year[time], 1, 1) + datetime.timedelta(dataDict.doy[time] - 1))
indexed_df = dataDict.set_index(pd.to_datetime(timestamp))
dataDict[dataDict == -9999] = np.nan
dailyData = dataDict
```

```python
# Ecosystem response binner

def siteBinner(SITE, toBin, binBy, numberOfBins, dailyData):
    testdf = dailyData[dailyData['SITE'] == SITE]
    testdf[testdf == -9999] = np.nan
    testdf.SW_IN_F.plot()
    toBin = toBin
    binby = binBy
    numberOfBins = 25
    binmin = testdf[binby].min()
    binmax = testdf[binby].max()
    binstep = (binmax - binmin) / numberOfBins
    xbinvec = []
    ybinvec = []
    binobs = []
    toperr = []
    bottomerr = []

    for binidx in arange(0,numberOfBins):
        binmean = nanmean(testdf[toBin][(testdf[binby] > (binstep * binidx)) & (testdf[binby] < (binstep * binidx + 1))])
        #print testdf[binby]
        #print testdf[toBin][(testdf[binby] > (binstep * binidx)) & (testdf[binby] < (binstep * binidx + 1))]
        binvar = (nanvar(testdf[toBin][(testdf[binby] > (binstep * binidx)) & (testdf[binby] < (binstep * binidx + 1))]))**(1/2)
        binob = len(testdf[toBin][(testdf[binby] > (binstep * binidx)) & (testdf[binby] < (binstep * binidx + 1))])
        if binob > 20:
            xbinvec.append(binstep*binidx)
            ybinvec.append(binmean)
            toperr.append((6*binvar/binob) + binmean)
            bottomerr.append(binmean - (6*binvar/binob))
   
    outputDict = pd.DataFrame(dict(SITE = SITE, VAR = binBy, xdata = xbinvec, ydata = ybinvec, 
                                   toperror = toperr, bottomerror = bottomerr))
    return outputDict

years = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]

for thisyear in years:
    for site in order:
        dailydDataY = dailyData[dailyData.year == thisyear]
        binnedData = binnedData.append(YEAR = year, siteBinner(site, 'FC_F', 'TA_F', 25, dailyDataY))
```
