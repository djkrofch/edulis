```python
import numpy
import scipy
import pyprosail

# Load required packages and standardize plot output a bit
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
import seaborn as sns

rcParams['axes.labelsize'] = 20
rcParams['axes.titlesize'] = 20
rcParams['xtick.labelsize'] = 15
rcParams['ytick.labelsize'] = 15
rcParams['legend.fontsize'] = 20
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Computer Modern Roman']
rcParams['text.usetex'] = False
```

```python
NDVI_r = []
NDVI_b = []
LAI = []
index = []
idx = 0
NDVIs = pd.DataFrame(dict(NDVI_red = NDVI_r, NDVI_blue = NDVI_b, LAI = LAI), index = index)


# Import a camera model, from a typical 3 band sensor

for LAI in arange(0,3, 0.1):
    index = [idx]
    spectra = pyprosail.run(1.5, 40, 8, 0, 0.01, 0.009, 1, LAI, 0.01, 30, 10, 0,0,pyprosail.Spherical)
    wavenumber = spectra[:,0]
    reflectance = spectra[:,1]

    blue = [.450, .520]
    green = [.520, .600]
    red = [.630, .690]
    nir = [.760, .900]

    blue_ref = mean(spectra[(wavenumber >= blue[0]) & (wavenumber < blue[1]), 1])
    green_ref = mean(spectra[(wavenumber >= green[0]) & (wavenumber < green[1]), 1])
    red_ref = mean(spectra[(wavenumber >= red[0]) & (wavenumber < red[1]), 1])
    nir_ref = mean(spectra[(wavenumber >= nir[0]) & (wavenumber < nir[1]), 1])

    NDVI_b = (nir_ref - blue_ref) / (blue_ref + nir_ref)
    NDVI_r = (nir_ref - red_ref) / (red_ref + nir_ref)

    NDVIs = NDVIs.append(pd.DataFrame(dict(NDVI_red = NDVI_r, NDVI_blue = NDVI_b, LAI = LAI), 
                                      index = index))
    idx = idx + 1
plot(wavenumber, reflectance)
sns.despine()
sns.set_style("white")
xlabel('wavelength')
ylabel('reflectance')
```

```python
fig, (ax1) = plt.subplots(1,1, figsize = (5,5))

NDVIs.plot('LAI', 'NDVI_blue', ax = ax1)
NDVIs.plot('LAI', 'NDVI_red', ax = ax1)
legend(['NDVI_b','NDVI_r'], loc = 'best')
```
