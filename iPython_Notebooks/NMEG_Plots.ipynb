```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

gcolour=[0.9,0.5, 0.0];
ngcolour=[0.9, 0.6, 0.0];
scolour=[0.6, 0.2, 0];
jcolour=[0.25, 1.0, 0.0];
pjcolour=[0.0, 0.5, 0.0];
pjgcolour=[0.2, 0.5, 0.0];
pcolour=[0.5, 0.5, 1.0];
mcolour=[0.0, 0.0, 0.6];
siteColors = [gcolour, scolour, jcolour, pjcolour, pcolour, mcolour]
```

```python
dataDir = '/home/nikko/Research/Data/TowerData/AmerifluxFiles/Products/'
dailyFileName = 'AllAmfluxData_Daily_Reichstein.csv'
dailyFileName_old = 'Old/AllAmfluxData_WithGaps_Filled_Daily_Reich.txt'

dailyData = pd.read_csv(dataDir + dailyFileName)
dailyData2 = pd.read_csv(dataDir + dailyFileName_old, sep = ' ')

timestamp = []
idx = 0
import datetime
for time in arange(0,dailyData.YEAR.size):
    #print time
    idx = idx + 1
    timestamp.append(datetime.datetime(int(dailyData.YEAR[time]), 1, 1) + datetime.timedelta(int(dailyData.DOY[time]) - 1))
indexed_df = dailyData.set_index(pd.to_datetime(timestamp))

timestamp = []
idx = 0
import datetime
for time in arange(0,dailyData2.YEAR.size):
    #print time
    idx = idx + 1
    timestamp.append(datetime.datetime(int(dailyData2.YEAR[time]), 1, 1) + datetime.timedelta(int(dailyData2.DOY[time]) - 1))

indexed_df2 = dailyData2.set_index(pd.to_datetime(timestamp))
```

```python
site = 'Mpj'
pjdata = dailyData[dailyData.SITE == site]
num_plots = unique(pjdata.YEAR).size
fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(1,7, 
                                                        sharey = True,
                                                        figsize = (21,3))
subplots_adjust(wspace = 0)
axs = fig.axes
idx = 0

for year in unique(pjdata.YEAR):
    yearlyData = pjdata[pjdata.YEAR == year]
    ax = axs[idx]
    yearlyData.plot(ax = ax, kind = 'line', x = 'DOY', y = 'RE')
    ax.set_xlim([1, 365]) 
    if idx % 2 != 0:
        ax.set_xticks([])
        ax.set_xlabel('')
        
    idx = idx + 1
    
```

```python
siteColors_p = list(siteColors)

gg = sns.factorplot("Site", "value", "NEE Component", melt_C, kind="box",
                   x_order=order,aspect=(1.45), size = 15)
sns.set_context("poster")
sns.set_style("ticks")

gg.despine(offset=10, trim=True);
gg.set_axis_labels("Site", "gC d$^{-1}$")
yticks(fontsize = 20)
xticks(fontsize = 20)
xlabel('Site',fontsize = 20)

savefig('GPP_Frac_Box.tif')

```
