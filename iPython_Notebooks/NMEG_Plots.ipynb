```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataDir = '/home/nikko/Research/Data/TowerData/AmerifluxFiles/Products/'
dailyFileName = 'AllAmfluxData_Daily_Reichstein.csv'
dailyData = pd.read_csv(dataDir + dailyFileName)
```

```python
site = 'Mpj'
pjdata = dailyData[dailyData.SITE == site]
num_plots = unique(pjdata.YEAR).size
fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(1,7, 
                                                        sharey = True,
                                                        figsize = (21,3))
subplots_adjust(wspace = 0)
axs = fig.axes
idx = 0

for year in unique(pjdata.YEAR):
    yearlyData = pjdata[pjdata.YEAR == year]
    ax = axs[idx]
    yearlyData.plot(ax = ax, kind = 'line', x = 'DOY', y = 'RE')
    ax.set_xlim([1, 365]) 
    if idx % 2 != 0:
        ax.set_xticks([])
        ax.set_xlabel('')
        
    idx = idx + 1
    
```

```python
order = list(['Seg','Sen','Ses','Wjs','Mpj','Mpj','Vcp','Vcm'])
sns.set_style("ticks")

f, ax = plt.subplots()
sns.violinplot(dailyData.TA_mean.dropna(), 
               dailyData.SITE.dropna(), order = order,
               color="coolwarm_r", lw = 2)

sns.despine(offset=10, trim=True);
sns.set_context("poster")
xlabel('Site')
ylabel('Mean Air Temp (deg C)')

f, ax = plt.subplots()
sns.violinplot(dailyData.FC.dropna(), 
               dailyData.SITE.dropna(), order = order,
               color="coolwarm_r", lw = 2)
sns.set_context("poster")
sns.despine(offset=10, trim=True);

factorData = pd.DataFrame(dict(Site = dailyData.SITE.dropna(),
                               RE = dailyData.RE.dropna(),
                               GPP = dailyData.GPP.dropna()))
melt_C = pd.melt(factorData, id_vars='Site', value_vars=['RE','GPP'], var_name='NEE Component')

```

```python
gg = sns.factorplot("Site", "value", "NEE Component", melt_C, kind="box",
                   x_order=order,aspect=(1.45), size =)
sns.set_context("poster")
sns.set_style("ticks")

gg.despine(offset=10, trim=True);
gg.set_axis_labels("Site", "gC d$^{-1}$")
```
