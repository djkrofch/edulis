<h3>
Function definition
</h3>
<h4>
<p>
Here I define a few simple functions to operate on the coregistered, atmospherically corrected RapidEye data. The general workflow here is to generate the file list for a given working directory, and operate over each file in the directory, getting information about the image from the file name. Briefly, the functions below: <p>
1) calculate NDVI and NDRE for each input image, and save those raster outputs, and 
<p>
2) generate zonal stats on each of those output VI rasters, by looping through a zonal stats function once for each feature in the input shapefile (twice in this case, once for control and once for girdle). 
<p>The most complex part of all of this is generating the output dictionary in a meaningful way, that can be easily converted to CSV later on.
</h4>

```python
# Generator function to produce file names
def files(path):
    for file in os.listdir(path):
        if os.path.isfile(os.path.join(path, file)):
            yield file

# Function to produce rasters of NDVI and NDRE from the time series of RapidEye
# TOA-DOS imagery. 
def calculate_VI(input_value_raster):
    # Open the data
    raster = gdal.Open(input_value_raster)
    
    # Parse the file name to extract the date
    fname = (input_value_raster).split('/')[6]
    geo = raster.GetGeoTransform()
    prj = raster.GetProjection()
    
    # RapidEye has 5 spectral bands, read each as an array
    # Some bands not used at this point **
    Blue = numpy.array(raster.GetRasterBand(1).ReadAsArray(), dtype = float)
    Green = numpy.array(raster.GetRasterBand(2).ReadAsArray(), dtype = float)
    Red = numpy.array(raster.GetRasterBand(3).ReadAsArray(), dtype = float)
    RedEdge = numpy.array(raster.GetRasterBand(4).ReadAsArray(), dtype = float)
    NIR = numpy.array(raster.GetRasterBand(5).ReadAsArray(), dtype = float)
    
    # Calculate the indices of interest (NDVI, and NDRE)
    NDVI = (NIR - Red) / (NIR + Red)
    NDRE = (RedEdge - Red) / (RedEdge + Red)

    # Get the dimensions from NIR band as a template for the output image shape
    [cols,rows] = NIR.shape
    
    # Specify the naming convention for output files
    outNDVI = 'NDVI_' + fname
    outNDRE = 'NDRE_' + fname
    
    # Create the file, using the information from the original file
    outdriver = gdal.GetDriverByName("GTiff")
    outndvi   = outdriver.Create(str(outputDir + outNDVI), rows, cols, 1, gdal.GDT_Float32)
    outndre   = outdriver.Create(str(outputDir + outNDRE), rows, cols, 1, gdal.GDT_Float32)

    # Write the array to the file, which is the original array in this example
    outndvi.GetRasterBand(1).WriteArray(NDVI)
    outndre.GetRasterBand(1).WriteArray(NDRE)

    # Georeference the image
    outndvi.SetGeoTransform(geo)
    outndre.SetGeoTransform(geo)

    # Write projection information
    outndvi.SetProjection(prj)
    outndre.SetProjection(prj)

# Function to produce a dictionary of raster values given a shapefile input
def zonal_stats(feat, input_zone_polygon, input_value_raster):

    # Open data
    raster = gdal.Open(input_value_raster)
    shp = ogr.Open(input_zone_polygon)
    lyr = shp.GetLayer()

    # Get raster georeference info
    transform = raster.GetGeoTransform()
    xOrigin = transform[0]
    yOrigin = transform[3]
    pixelWidth = transform[1]
    pixelHeight = transform[5]

    # Get extent of feat
    geom = feat.GetGeometryRef()
    if (geom.GetGeometryName() == 'MULTIPOLYGON'):
        count = 0
        pointsX = []; pointsY = []
        for polygon in geom:
            geomInner = geom.GetGeometryRef(count)    
            ring = geomInner.GetGeometryRef(0)
            numpoints = ring.GetPointCount()
            for p in range(numpoints):
                    lon, lat, z = ring.GetPoint(p)
                    pointsX.append(lon)
                    pointsY.append(lat)    
            count += 1
    elif (geom.GetGeometryName() == 'POLYGON'):
        ring = geom.GetGeometryRef(0)
        numpoints = ring.GetPointCount()
        pointsX = []; pointsY = []
        for p in range(numpoints):
                lon, lat, z = ring.GetPoint(p)
                pointsX.append(lon)
                pointsY.append(lat)

    else:
        sys.exit()

    xmin = min(pointsX)
    xmax = max(pointsX)
    ymin = min(pointsY)
    ymax = max(pointsY)

    # Specify offset and rows and columns to read
    xoff = int((xmin - xOrigin)/pixelWidth)
    yoff = int((yOrigin - ymax)/pixelWidth)
    xcount = int((xmax - xmin)/pixelWidth)+1
    ycount = int((ymax - ymin)/pixelWidth)+1

    # Create memory target raster
    target_ds = gdal.GetDriverByName('MEM').Create('', xcount, ycount, gdal.GDT_Byte)
    target_ds.SetGeoTransform((
        xmin, pixelWidth, 0,
        ymax, 0, pixelHeight,
    ))

    # Create for target raster the same projection as for the value raster
    raster_srs = osr.SpatialReference()
    raster_srs.ImportFromWkt(raster.GetProjectionRef())
    target_ds.SetProjection(raster_srs.ExportToWkt())

    # Rasterize zone polygon to raster
    gdal.RasterizeLayer(target_ds, [1], lyr, burn_values=[1])

    # Read raster as arrays
    banddataraster = raster.GetRasterBand(1)
    dataraster = banddataraster.ReadAsArray(xoff, yoff, xcount, ycount).astype(numpy.float)

    bandmask = target_ds.GetRasterBand(1)
    datamask = bandmask.ReadAsArray(0, 0, xcount, ycount).astype(numpy.float)

    # Mask zone of raster
    zoneraster = numpy.ma.masked_array(dataraster,  numpy.logical_not(datamask))

    # Calculate statistics of zonal raster
    return numpy.mean(zoneraster)


# Helper function to loop through the zonal stats function
# specifically, this function loops through each feature within the
# input shapefile
def loop_zonal_stats(input_zone_polygon, input_value_raster):

    shp = ogr.Open(input_zone_polygon)
    lyr = shp.GetLayer()
    featList = range(lyr.GetFeatureCount())
    statDict = {}
    for FID in featList:
        feat = lyr.GetFeature(FID)
        meanValue = zonal_stats(feat, input_zone_polygon, input_value_raster)
        statDict[FID] = meanValue
    return statDict
```

<h4>
Calculate NDVI and NDRE for each scene, by calling calculate_VI, defined above
</h4>

```python
filelist = files(dataDir)
for file in filelist:
    name = file
    calculate_VI(dataDir + name)
```

<h4>
Define an empty dictionary, and fill it with the output from loop_zonal_stats, operating on the scene VI.
</h4>

```python
filelist = []
d = {}
dlist = []
check = files(outputDir)
for names in check:
    filelist.append(names)
    
for names in filelist:
    name = names
    index = (name).split('_')[0]
    date = (name).split('_')[1].split('.')[0]
    imgname = outputDir + name
    shapename = shapeDir + 'PJ_Fetch.shp'

    d=dict(filename = name,index = index,date = date,
      controlVI= loop_zonal_stats(shapename, imgname)[0],
      girdleVI = loop_zonal_stats(shapename, imgname)[1])
    dlist.append(d)
```

<h4>
Write the dictionary to CSV so that we can move on to combining the RE and ETM data in the R script (RE_Linear_Modeling.R), and get on with the linear model generation.
</h4>

```python
import csv
toCSV = dlist
keys = toCSV[0].keys()
with open('RapidEye_VI.csv', 'wb') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(toCSV)
```
