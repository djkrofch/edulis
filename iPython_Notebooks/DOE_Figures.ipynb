```python
dataDir = '/home/nikko/Research/Data/TowerData/AmerifluxFiles/Products/'
dailyFileName = 'AllAmfluxData_Daily_Reichstein.csv'
min30filename = 'AllAmfluxData_30min_Reichstein.csv'
fluxdata = pd.read_csv(dataDir + min30filename)
dailydata = pd.read_csv(dataDir + dailyFileName)
grounddata = pd.read_csv('PJC_Tree_Mortality_Survey_Data.csv')

pjc = fluxdata[fluxdata.SITE == 'Mpj']
pjcD = dailydata[dailydata.SITE == 'Mpj']
```

```python
HRMINs = [0, 30, 100, 130, 200, 230, 300, 330, 400, 
          430, 500, 530, 600, 630, 700, 730, 800, 
          830, 900, 930, 1000, 1030, 1100, 1130, 1200,
          1230, 1300, 1330, 1400, 1430, 1500, 1530, 
          1600, 1630, 1700, 1730, 1800, 1830, 1900,
          1930, 2000, 2030, 2100, 2130, 2200, 2230,
          2300, 2330, 2400]

binnedY = []
binnedX = []
binnedYear = []
idx = 0
var1 = 'FC'
years = [2009, 2010, 2011, 2012, 2013, 2014]
for year in years:
    pjcy = pjc[pjc.YEAR == year]
    for hour in arange(0,24):
        binnedYear.append(year)
        binnedY.append(pjcy[var1][(pjcy.HRMIN >= HRMINs[idx]) & (pjcy.HRMIN < HRMINs[idx + 2])].mean())

        binnedX.append(hour)
        idx = idx + 2
    idx = 0
hourlyData = pd.DataFrame(dict(Year = binnedYear, Hour = binnedX, Data = binnedY))    
```

```python
fig, (ax1) = plt.subplots(1,1, figsize = (5,5))
axs = fig.axes
idx = 0

for year in years:
    hourlyData[hourlyData.Year == year].plot(ax = ax1,x = 'Hour', y = 'Data', label = year)
    ax1.legend(years, fontsize = 12)
    ax1.grid(False)
    ax1.set_xlabel('Hour', fontsize = 20)
    ax1.set_ylabel('GPP ($\mu$mol C m$^{-2}$ s$^{-1}$)', fontsize = 20)
    ax1.xaxis.set_tick_params(labelsize=20)
    ax1.yaxis.set_tick_params(labelsize=20)

fig.tight_layout()
sns.set_style("white")
savefig('FC_Diurnal.tif')
```
